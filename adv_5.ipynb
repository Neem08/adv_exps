```r
# Install necessary libraries if not already installed
install.packages(c("ggplot2", "plotly", "wordcloud", "scatterplot3d"))

# Load libraries
library(ggplot2)
library(plotly)
library(wordcloud)
library(scatterplot3d)

# Load the dataset (make sure to set the correct path)
housing_data <- read.csv("housing.csv")
```

```r
# Install the required packages
install.packages("tm")
install.packages("slam")
```

```r
# Word chart to show the frequency of each type of ocean proximity
wordcloud(housing_data$ocean_proximity, min.freq = 1, colors = brewer.pal(8, "Dark2"))

```

```r
# Box plot of Median House Value by Ocean Proximity
ggplot(housing_data, aes(x = ocean_proximity, y = median_house_value, fill = ocean_proximity)) +
  geom_boxplot() +
  labs(title = "Median House Value by Ocean Proximity", x = "Ocean Proximity", y = "Median House Value") +
  theme_minimal()
```

```r
# Violin plot of Median Income by Housing Median Age
ggplot(housing_data, aes(x = as.factor(housing_median_age), y = median_income, fill = as.factor(housing_median_age))) +
  geom_violin() +
  labs(title = "Median Income by Housing Median Age", x = "Housing Median Age", y = "Median Income") +
  theme_minimal()
```

```r
# Box plot of Median House Value by Ocean Proximity
ggplot(housing_data, aes(x = ocean_proximity, y = median_house_value, fill = ocean_proximity)) +
  geom_boxplot() +
  labs(title = "Median House Value by Ocean Proximity", x = "Ocean Proximity", y = "Median House Value") +
  theme_minimal()
```

```r
# Linear regression plot for Median House Value vs. Median Income
ggplot(housing_data, aes(x = median_income, y = median_house_value)) +
  geom_point() +
  geom_smooth(method = "lm", color = "blue") +
  labs(title = "Linear Regression of Median House Value vs Median Income", x = "Median Income", y = "Median House Value") +
  theme_minimal()
```

```r
# Nonlinear regression plot using LOESS (locally estimated scatterplot smoothing)
ggplot(housing_data, aes(x = population, y = median_house_value)) +
  geom_point() +
  geom_smooth(method = "loess", color = "red") +
  labs(title = "Nonlinear Regression of Median House Value vs Population", x = "Population", y = "Median House Value") +
  theme_minimal()
```

```r
# 3D scatter plot of Median Income, Housing Median Age, and Median House Value
scatterplot3d(housing_data$median_income, housing_data$housing_median_age, housing_data$median_house_value,
              xlab = "Median Income", ylab = "Housing Median Age", zlab = "Median House Value",
              color = "blue", pch = 16)
```

```r
# Jitter plot to reduce overlapping of points for Total Rooms vs. Median Income
ggplot(housing_data, aes(x = median_income, y = total_rooms)) +
  geom_jitter(width = 0.2, height = 0.2, color = "darkgreen") +
  labs(title = "Jitter Plot of Total Rooms vs Median Income", x = "Median Income", y = "Total Rooms") +
  theme_minimal()
```
